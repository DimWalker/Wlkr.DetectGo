```python
# ä¸‹è½½è¡—æ™¯
!mkdir -p /content/Wlkr.DetectGo/output
%cd /content/Wlkr.DetectGo/output
!unzip -qo /content/drive/MyDrive/go_board_det_yolov8/labelmefacade-master.zip

# åˆ’åˆ†æ•°æ®é›†
%cd /content/Wlkr.DetectGo/output
# åˆ é™¤æ—§æ•°æ®
!rm -rf /content/Wlkr.DetectGo/output/scene_images_train
!rm -rf /content/Wlkr.DetectGo/output/scene_images_eval


import os
import shutil
from random import shuffle


pre_path = "/content"

# æŒ‡å®šæ•°æ®é›†ç›®å½•å’Œå­ç›®å½•
dataset_dir = os.path.join(pre_path, "Wlkr.DetectGo/output/labelmefacade-master/images")
train_dir = os.path.join(pre_path, "Wlkr.DetectGo/output/scene_images_train")
val_dir = os.path.join(pre_path, "Wlkr.DetectGo/output/scene_images_eval")

# ç¡®ä¿è®­ç»ƒé›†å’Œè¯„ä¼°é›†ç›®å½•å­˜åœ¨
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# è·å–æ‰€æœ‰å›¾ç‰‡æ–‡ä»¶
images = [f for f in os.listdir(dataset_dir) if os.path.isfile(os.path.join(dataset_dir, f))]

# æ‰“ä¹±å›¾ç‰‡åˆ—è¡¨ä»¥éšæœºåŒ–
shuffle(images)

# è®¡ç®—åˆ’åˆ†ç‚¹
split_point = int(len(images) * 0.8)  # 80% ç”¨äºè®­ç»ƒé›†

# åˆ†å‰²æ•°æ®é›†
train_images = images[:split_point]
val_images = images[split_point:]

# å°†å›¾ç‰‡ç§»åŠ¨åˆ°ç›¸åº”çš„ç›®å½•
for image in train_images:
    shutil.move(os.path.join(dataset_dir, image), os.path.join(train_dir, image))

for image in val_images:
    shutil.move(os.path.join(dataset_dir, image), os.path.join(val_dir, image))

print(f"æ•°æ®é›†å·²åˆ’åˆ†ï¼š\nè®­ç»ƒé›†åŒ…å« {len(train_images)} å¼ å›¾ç‰‡ï¼Œä½äº {train_dir}\nè¯„ä¼°é›†åŒ…å« {len(val_images)} å¼ å›¾ç‰‡ï¼Œä½äº {val_dir}")

# åˆ é™¤è¡—æ™¯æ— ç”¨æ–‡ä»¶
!rm -rf /content/Wlkr.DetectGo/output/labelmefacade-master
```
